```python
def atividades_ano_mes(df_teste, ano, mes):
    colunas_desejadas = ['data_ano', 'data_mes', 'dia_semana', 'Activity Name',	'Activity Type', 'Elapsed Time',	'Distance', 'Calories', 'tempo_min']

    novo_df = df_teste[colunas_desejadas]
    novo_df = novo_df[(novo_df['data_ano'] == ano) & (novo_df['data_mes'] == mes)]
    
    return novo_df
```
